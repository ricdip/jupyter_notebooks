```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("seaborn")
```

```python
x = np.linspace(-np.pi,np.pi,100)
```

```python
plt.plot(x,np.sin(x), linestyle='-')
plt.plot(x,np.cos(x), linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x), color='blue')
ax[1].plot(x, np.cos(x), color='red')
```

```python
plt.title('sin/cos functions')
plt.plot(x,np.sin(x), linestyle='-', label='sin(x)')
plt.plot(x,np.cos(x), linestyle='--', label='cos(x)')
plt.legend()
plt.show()
```

```python
x = np.linspace(-np.pi,np.pi, 20)
plt.scatter(x,np.sin(x), color='black')
```

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
```

```python
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X,Y,Z, colors='blue')
```

```python
from mpl_toolkits import mplot3d
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
```

```python
X,Y = np.meshgrid(x, y)
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='Blues')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.view_init(60,35)
ax.set_title('plotting 3d')
fig
```
